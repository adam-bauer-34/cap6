# Parsing Burke _et al_., 2018 damage function data

David supplied me with the raw output of the Burke _et al_., 2018 data for damage functions. They're all in one file. In this notebook I parse the data into their own files based on SSP.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse_data(filename, year, check_plots=True, 
               path='/data/keeling/a/adammb4/ClimateEcon/ez-climate/TCREZClimate/data/'):
    """Parse Burke et al., reproduction data.
    
    This function splits up one file of data into five files of data, separated by SSP.
    
    Parameters
    ----------
    filename: string
        name of file
        
    year: int
        year of data (for saved file name)
    
    check_plots: bool (default True)
        make check plots?
        
    path: string (default Adam's data folder on Keeling)
        path to data/to saving directory
    
    Returns
    -------
    if check_plots:
        makes check plots
    """
    
    # import data (first line) and ssps (second line)
    df_t_d = np.genfromtxt(''.join([path, filename]), 
                                delimiter=',', usecols=(2,3), skip_header=1)
    df_baselines = np.genfromtxt(''.join([path, filename]), 
                                delimiter=',', usecols=0, dtype="|U5", skip_header=1)
    
    # get indicies equal to each ssp
    ssp1_inds = np.where(df_baselines == '"ssp1')
    ssp2_inds = np.where(df_baselines == '"ssp2')
    ssp3_inds = np.where(df_baselines == '"ssp3')
    ssp4_inds = np.where(df_baselines == '"ssp4')
    ssp5_inds = np.where(df_baselines == '"ssp5')
    
    # parse temperature and damage data
    df_t, df_d = df_t_d.T
    
    ssp1_t = df_t[ssp1_inds] + 0.8
    ssp2_t = df_t[ssp2_inds] + 0.8
    ssp3_t = df_t[ssp3_inds] + 0.8
    ssp4_t = df_t[ssp4_inds] + 0.8
    ssp5_t = df_t[ssp5_inds] + 0.8

    ssp1_d = -1 * df_d[ssp1_inds] * 0.01
    ssp2_d = -1 * df_d[ssp2_inds] * 0.01
    ssp3_d = -1 * df_d[ssp3_inds] * 0.01
    ssp4_d = -1 * df_d[ssp4_inds] * 0.01
    ssp5_d = -1 * df_d[ssp5_inds] * 0.01
    
    # make new arrays, save them 
    # ssp1
    t_d_ssp1 = np.zeros_like((ssp1_t, ssp1_d))
    t_d_ssp1[0] = ssp1_t
    t_d_ssp1[1] = ssp1_d
    t_d_ssp1 = t_d_ssp1.T

    np.savetxt(''.join([path, 'ssp1_', str(year), '_t_d.csv']), t_d_ssp1, delimiter=',')
    
    # ssp2
    t_d_ssp2 = np.zeros_like((ssp2_t, ssp2_d))
    t_d_ssp2[0] = ssp2_t
    t_d_ssp2[1] = ssp2_d
    t_d_ssp2 = t_d_ssp2.T

    np.savetxt(''.join([path, 'ssp2_', str(year), '_t_d.csv']), t_d_ssp2, delimiter=',')
    
    # ssp3
    t_d_ssp3 = np.zeros_like((ssp3_t, ssp3_d))
    t_d_ssp3[0] = ssp3_t
    t_d_ssp3[1] = ssp3_d
    t_d_ssp3 = t_d_ssp3.T

    np.savetxt(''.join([path, 'ssp3_', str(year), '_t_d.csv']), t_d_ssp3, delimiter=',')
    
    # ssp4
    t_d_ssp4 = np.zeros_like((ssp4_t, ssp4_d))
    t_d_ssp4[0] = ssp4_t
    t_d_ssp4[1] = ssp4_d
    t_d_ssp4 = t_d_ssp4.T

    np.savetxt(''.join([path, 'ssp4_', str(year), '_t_d.csv']), t_d_ssp4, delimiter=',')
    
    # ssp5
    t_d_ssp5 = np.zeros_like((ssp5_t, ssp5_d))
    t_d_ssp5[0] = ssp5_t
    t_d_ssp5[1] = ssp5_d
    t_d_ssp5 = t_d_ssp5.T

    np.savetxt(''.join([path, 'ssp5_', str(year), '_t_d.csv']), t_d_ssp5, delimiter=',')
    
    # make check plots if so desired
    if check_plots:
        fig, ax = plt.subplots(1,5, figsize=(20, 4))

        ax[0].scatter(ssp1_t, ssp1_d)
        ax[1].scatter(ssp2_t, ssp2_d)
        ax[2].scatter(ssp3_t, ssp3_d)
        ax[3].scatter(ssp4_t, ssp4_d)
        ax[4].scatter(ssp5_t, ssp5_d)
        
        for i in range(5):
            ax[i].set_xlabel("T anom.")
        
        ax[0].set_ylabel("Damage (% GDP)")
```

```python
parse_data('FigED6_data.csv', 2099)
```
