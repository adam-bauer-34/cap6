# Cost function fitting

Using the data from IPCC WG3, we fit the cost curve to a second order polynomial. This is the replacement of the McKinsie curve, and thus, the `cost` function in `TCREZClimate`.

NOTE: This curve assumes that the mitigation potentials across sectors are addivitve. The IPCC warns that this is not precisely true, but it's unclear right now how this lack of additivity would be accounted for.

```python
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib as mpl
from scipy.optimize import curve_fit

# plotting parameters
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
color_list = ['#000000', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#CC79A7']
marker_list = ['o', 's', 'P', '+', 'D', 'v', '3']
markersize = 6
linestyle_list = ['solid', 'dashed', 'dashdot', 'dotted'] * 2
linewidth = 2
fontsize = 22
labelsize = 20
figsize = (16, 10)
node_times = [2020, 2030, 2060, 2100, 2200, 2300]
x_label = 'Years'
```

This is the data from IPCC. The IPCC gives cost ranges for each mitigation potential; namely
- $<$ \$0
- \\$0 - \$20
- \\$20 - \$50
- \\$50 - \$100
- \\$100 - \$200

We make a few cost curves; one that is "optimistic," and assumes that the highest cost within a given range of mitigation potential is reached once the highest mitigation potential is acualized. For example, if the cost range \\$0 - \\$20 is being considered, and the technologies in this range have the potential to abate 10 GtCO$_2$-e, then it costs \\$0 to abate 0 GtCO$_2$-e for that range, and \\$20 to abate 10 GtCO$_2$-e. 

```python
cost_opt = np.array([0,0,20,50,100,200], dtype=np.float32)
cost_pes = np.array([0,20,50,100,200], dtype=np.float32)

mits = np.array([0,10.6, 9, 7.3, 9.9, 3.8], dtype=np.float32)
mits_range = np.zeros_like(mits)
for i in range(len(mits)):
    mits_range[i] = np.sum(mits[0:i+1])
    
mits_opt = mits_range
mits_pes = mits_range[:-1]
mits_mid = np.zeros_like(mits)
for i in range(1,len(mits)):
    mits_mid[i] = 0.5 * (mits_range[i-1] + mits_range[i])
    
mits_mid[1] = mits_range[1]
mits_opt, mits_pes, mits_mid
```

Fit second-order polynomial to the data.

```python
E_2030 = 34.734424 # SSP1
#E_2030 = 43.476063 # SSP2
#E_2030 = 52.847359 # SSP3
#E_2030 = 45.167913 # SSP4
#E_2030 = 55.296583 # SSP5
1 - (E_2030 - mits_opt) * E_2030**(-1)
```

```python
def f(x, a, b):
    return a * x**b

opt_fit = curve_fit(f, 1 - (E_2030 - mits_opt) * E_2030**(-1) + 1e-10, cost_opt)
pes_fit = curve_fit(f, 1 - (E_2030 - mits_pes) * E_2030**(-1) + 1e-10, cost_pes)
mid_fit = curve_fit(f, 1 - (E_2030 - mits_mid) * E_2030**(-1) + 1e-10, cost_opt)
opt_coeff, opt_power = opt_fit[0]
pes_coeff, pes_power = pes_fit[0]
mid_coeff, mid_power = mid_fit[0]

pes_coeff, pes_power
```

```python
mits_full_range = np.arange(0,1,0.01)
```

```python
fig, ax = plt.subplots(1, figsize=figsize)

#ax.scatter(1 - (E_2030 - mits_opt) * E_2030**(-1), cost_opt, label="Fit Method #1 Data", color=color_list[0], marker=marker_list[0])
ax.scatter(1 - (E_2030 - mits_pes) * E_2030**(-1), cost_pes, label="AR6 WGIII data", color=color_list[1], marker=marker_list[1])
#ax.scatter(1 - (E_2030 - mits_mid) * E_2030**(-1), cost_opt, label="Fit Method #1 Data", color=color_list[2], marker=marker_list[2])

#ax.plot(mits_full_range, opt_coeff * mits_full_range**opt_power, label="Fit Method #1 Curve", 
#        color=color_list[0], linewidth=linewidth, markersize=markersize, linestyle=linestyle_list[0])
ax.plot(mits_full_range, pes_coeff * mits_full_range**pes_power, label="AR6 WGIII curve fit",
       color=color_list[1], linewidth=linewidth, markersize=markersize, linestyle=linestyle_list[1])
#ax.plot(mits_full_range, mid_coeff * mits_full_range**mid_power, label="Fit AR6 Curve",
#       color=color_list[2], linewidth=linewidth, markersize=markersize, linestyle=linestyle_list[2])
ax.plot(mits_full_range, 1.21 * 314.32 * mits_full_range**2.413, label="McKinsey Cost Curve", color=color_list[3],
        linewidth=linewidth, markersize=markersize, linestyle=linestyle_list[3])
ax.legend(fontsize=fontsize, frameon=False)
ax.set_xlabel("Mitigation Potential (Fraction of 2030 emissions)", fontsize=fontsize)
ax.set_ylabel("Cost (\$ 2020 USD)", fontsize=fontsize)
ax.tick_params(axis="both", labelsize=labelsize)
```
